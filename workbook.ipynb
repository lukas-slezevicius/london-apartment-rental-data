```python
import folium
import csv
import geopandas as gpd
from folium.plugins import MarkerCluster
```

```python
m = folium.Map(location=[51.5072, -0.1276], zoom_start=10, tiles="OpenStreetMap")
marker_cluster = MarkerCluster().add_to(m)
with open("data/find-my-area-co-uk.csv", "r", newline="") as f:
    reader = csv.DictReader(f)
    for row in reader:
        area_name = row["name"]
        lat, lon = float(row["latitude"]), float(row["longitude"])
        folium.Marker([lat, lon], tooltip=area_name).add_to(marker_cluster)
# folium.GeoJson("data/london_boroughs.json").add_to(m)
# folium.GeoJson("data/ulez.json").add_to(m)
# folium.GeoJson("data/course_m25_boundary.json").add_to(m)
# ward_shape = gpd.read_file("data/London-wards-2018_ESRI/London_Ward.shp")
# folium.GeoJson(data=ward_shape["geometry"]).add_to(m)
# folium.GeoJson("data/tfl_lines.json").add_to(m)

```
